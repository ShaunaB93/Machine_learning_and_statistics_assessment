```python
import keras as kr
import pandas as pd
import sklearn.preprocessing as pre
import sklearn.model_selection as mod
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

from sklearn import datasets
from sklearn.utils import Bunch

from keras import models, layers
```

```python
boston = datasets.load_boston()
type(boston)
```

```python
df = pd.DataFrame(data=boston['data'], columns=boston['feature_names'])
df['MEDV'] =boston['target']
medv = df['MEDV'].copy()
del df['MEDV']
df = pd.concat((medv, df), axis=1)

df
```

```python
inputs = df[['CRIM', 'CHAS', 'RM', 'DIS', 'TAX', 'AGE', 'LSTAT']]
inputs
```

```python
outputs = df[['MEDV']]
outputs
```

## Neural Network

```python
x_train, x_test, y_train, y_test = mod.train_test_split(inputs, outputs, test_size=0.4)
```

```python
m = kr.models.Sequential()

m.add(kr.layers.Dense(30, input_dim=7, activation="relu"))
#m.add(kr.layers.Dense(20, activation="relu"))
#m.add(kr.layers.Dense(5, activation="relu"))
m.add(kr.layers.Dense(1, activation="relu"))

m.compile(loss="mean_squared_error", optimizer="adam")
```

```python
m.fit(x_train,  y_train, epochs=500, batch_size=20)
```

```python
m.predict(x_test).round().T
```

```python
y_test.as_matrix().astype(np.float32)
```

```python
m.evaluate(x_test, y_test)
```

### Whiten the data

```python
import sklearn.decomposition as dec
```

```python
pca = dec.PCA(n_components=7, whiten=True)
pca.fit(inputs)
inputsWhite = pd.DataFrame(pca.transform(inputs), columns=inputs.columns)
inputsWhite 
```

```python
model = kr.models.Sequential()

model.add(kr.layers.Dense(20, input_dim=7, activation="relu"))
model.add(kr.layers.Dense(10, activation="linear"))
#model.add(kr.layers.Dense(10, activation="linear"))
model.add(kr.layers.Dense(1, activation="linear"))

model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
model.fit(inputsWhite, outputs, epochs=500, batch_size=10)
```

```python
model.predict(inputsWhite)
```

```python
model.predict(inputsWhite).T - outputs.as_matrix()
```

```python
np.sum((model.predict(inputsWhite).T - outputs.as_matrix())**2)
```

```python
np.sqrt(np.sum((model.predict(inputsWhite).T - outputs.as_matrix())**2))
```

```python
m.evaluate(inputsWhite, outputs)
```
