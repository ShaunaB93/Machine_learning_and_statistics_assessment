```python
import keras as kr
import pandas as pd
import sklearn.preprocessing as pre
import sklearn.model_selection as mod
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

import sklearn

from sklearn import datasets
from sklearn.utils import Bunch

from keras import models, layers

import sklearn.decomposition as dec
```

```python
boston = datasets.load_boston()
type(boston)
```

```python
df = pd.DataFrame(data=boston['data'], columns=boston['feature_names'])
df['MEDV'] =boston['target']
medv = df['MEDV'].copy()
del df['MEDV']
df = pd.concat((medv, df), axis=1)

df
```

```python
#inputs = df.loc[:, df.columns != 'MEDV']
#inputs = df[['LSTAT', 'RM', 'PTRATIO', 'INDUS', 'TAX']]
inputs = df[['LSTAT', 'RM', 'PTRATIO', 'INDUS']]
inputs
```

```python
outputs = df[['MEDV']]
outputs
```

Neural network will predict the MEDV value based on the 13 other variables in the dataset

```python
model = kr.models.Sequential()

model.add(kr.layers.Dense(units=30, activation='relu', input_shape=(4,)))
model.add(kr.layers.Dense(units=50, activation='relu'))
model.add(kr.layers.Dense(units=20, activation='relu'))
model.add(kr.layers.Dense(units=1))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
inputs_train, inputs_test, outputs_train, outputs_test = mod.train_test_split(inputs, outputs, test_size=0.2)
```

```python
model.fit(inputs_train, outputs_train, epochs=400, batch_size=50)
```

```python
scaler = pre.StandardScaler()
scaler.fit(inputs_train)
x_train_scaled = scaler.transform(inputs_train)
x_test_scaled = scaler.transform(inputs_test)
```

```python
model.fit(x_train_scaled, outputs_train, epochs=400, batch_size=50)
```

```python
pca = dec.PCA(n_components=4, whiten=True)
pca.fit(inputs_train)
inputs_train_white = pd.DataFrame(pca.transform(inputs_train), columns=inputs.columns)
inputs_train_white
```

```python
model.fit(inputs_train_white, outputs_train, epochs=400, batch_size=50)
```

```python
scaler = pre.StandardScaler()
scaler.fit(inputs_train_white)
x_train_scaledW = scaler.transform(inputs_train_white)
x_test_scaledW = scaler.transform(inputs_train_white)
```

```python
model.fit(x_train_scaledW, outputs_train, epochs=400, batch_size=50)
```
