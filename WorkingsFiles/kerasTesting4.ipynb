```python
import keras as kr
import pandas as pd
import sklearn.preprocessing as pre
import sklearn.model_selection as mod
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

from sklearn import datasets
from sklearn.utils import Bunch

from keras import models, layers
```

```python
boston = datasets.load_boston()
type(boston)
```

```python
df = pd.DataFrame(data=boston['data'], columns=boston['feature_names'])
```

```python
df['MEDV'] =boston['target']
```

```python
medv = df['MEDV'].copy()
del df['MEDV']
df = pd.concat((medv, df), axis=1)
```

```python
df
```

```python
inputs = df[['CRIM', 'CHAS', 'RM', 'DIS', 'TAX', 'AGE', 'LSTAT']]
inputs
```

```python
outputs = df[['MEDV']]
outputs
```

```python
model = kr.models.Sequential()

model.add(kr.layers.Dense(20, input_dim=7, activation="linear"))
model.add(kr.layers.Dense(20, activation="linear"))
model.add(kr.layers.Dense(1, activation="linear"))

model.compile(loss="mean_squared_error", optimizer="adam")
```

```python
model.fit(inputs, outputs, epochs=20, batch_size=10)
```

```python
model.predict(inputs)
```

```python
(outputs - model.predict(inputs))
```
