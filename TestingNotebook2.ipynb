```python
# Importing library
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
```

```python
boston = load_boston()
```

```python
print(boston.data.shape)
```

## Attribute information in column order for the dataset loaded - 

- CRIM per capita crime rate by town
- ZN proportion of residential land zoned for lots over 25,000 sq.ft.
- INDUS proportion of non-retail business acres per town
- CHAS Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
- NOX nitric oxides concentration (parts per 10 million)
- RM average number of rooms per dwelling
- AGE proportion of owner-occupied units built prior to 1940
- DIS weighted distances to five Boston employment centres
- RAD index of accessibility to radial highways
- TAX full-value property-tax rate per 10,000 dollars
- PTRATIO pupil-teacher ratio by town
- B 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
- LSTAT Percentage lower status of the population
- MEDV Median value of owner-occupied homes in 1000â€™s dollars

```python
housingFeatures = boston.data
housingPrices = boston.target
```

```python
attributes = boston.feature_names #Takes the category from each column header
print(attributes)
```

```python
#Store in DataFrame for use in some descriptive functions

bostondf_data = pd.DataFrame(housingFeatures, columns = attributes) #Gives a dataframe of all of the attributes noted for the houses
bostondf_target = pd.DataFrame(housingPrices, columns =['MEDV']) #Median value of homes
#Concates both the data and target information
df_boston = pd.concat([bostondf_data, bostondf_target,], axis = 1) #Combines the attribute and house data in to a single dataframe 
```

```python
#Describes the details contained in the dataset
print(boston.DESCR)
```

```python
#Prints the number of features measured & tests contained in the datset
hFeatures = df_boston.shape[1]
hRecords = df_boston.shape[0]

print("The Number of Housing Features Recorded in the Dataset: ", hFeatures)
print("The Number of properties tested", hRecords)
```

```python
# Extracting specific stats from the describe table -

# Minimum house value - 
minPrice = housingPrices.min()
print("The minimum house price in the dataset is: ", minPrice)

# Maximum house value - 
maxPrice = housingPrices.max()
print("The maximum house price in the dataset is: ", maxPrice)

# Median house value - 
medPrice = np.median(housingPrices)
print("The median house price in the dataset is: ", medPrice)

# Standard deviation of house prices
stdDevPrice = housingPrices.std()
print("The standard deviation of the house prices in the dataset is: {0:.2f}".format(stdDevPrice))
```

```python
# Scatter plot of median house prices v no. of rooms
pl.title("Median House Price vs Number of Rooms")
sns.scatterplot(df_boston.MEDV, df_boston.RM)
pl.xlabel("House prices ($10,000 values)")
pl.ylabel("Average Number of Rooms")

```

```python
# Scatter plot of median house prices v crime rate by town
pl.title("Median House Price vs Crime Rate")
sns.scatterplot(df_boston.MEDV, df_boston.CRIM)
pl.xlabel("House prices ($10,000 values)")
pl.ylabel("Crime rate")
```
